```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=np.linspace(0,5,11)
x
y=x**2
y
```

```python
x
```

```python
plt.plot(x,y)
plt.title("X vs Y")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
plt.subplot(2,2,1)

plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)

plt.subplot(2,2,4)
plt.plot(x**2,x**3)
```

```python
fig=plt.figure(figsize=(10,10), dpi=500)
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2=fig.add_axes([0.3,0.4,0.5,0.5])
axis2.plot(x,y)
```
